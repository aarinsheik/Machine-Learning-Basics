Decision Tree

```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

```python
inputs = df.drop("salary_more_then_100k",axis='columns')
inputs
```

```python
target = df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_com = LabelEncoder() 
le_job = LabelEncoder()
le_deg = LabelEncoder()
```

```python
inputs['company_n'] = le_com.fit_transform( inputs['company'])
inputs['job_n'] = le_job.fit_transform( inputs['job'])
inputs['degree_n'] = le_deg.fit_transform( inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'] , axis='columns')
inputs_n
```

```python
from sklearn import tree
mod = tree.DecisionTreeClassifier( criterion='entropy')
```

```python
mod.fit(inputs_n,target)
```

```python
mod.score(inputs_n , target)
```

```python
mod.predict([[2,2,1]])
```

```python
mod.predict([[2,0,1]])
```
