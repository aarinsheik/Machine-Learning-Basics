# Principle Component Anaylsis

Principle Component Anaylsis :
Is a process of figuring out most important features or principal components that has the most impact on the target variable



when we call PCA( is_component=6 ) , it gives most important 6 columns in dataset , this helps us to eliminate the unwanted data from chosen dataset (It is a dimensional reduction technique )

```python
from sklearn.datasets import load_digits
import pandas as pd
```

```python
dataset = load_digits()
dataset.keys()
```

```python
dataset.data[0]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.gray()
plt.matshow( dataset.data[50].reshape(8,8))
```

```python
dataset.target[50]
```

```python
df = pd.DataFrame( dataset.data , columns=dataset.feature_names )
df
```

```python
df.describe()
```

```python
x = df
y = dataset.target
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scl = StandardScaler()
x_scl = scl.fit_transform(x)
x_scl
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

xtr , xts ,ytr , yts = train_test_split( x_scl , y , test_size=0.3 , random_state=30 )

lr = LogisticRegression()
lr.fit( xtr , ytr )
```

```python
lr.score(xts ,yts)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
x_pca = pca.fit_transform(x)
x_pca.shape

# initial ,it had 64 column. but now, it has only 29 (PCA reduced unwanted columns )
```

```python
x_pca
```

```python
pca.explained_variance_ratio_ 

# it shows 1st col as 14.8% of useful info and 2nd col - 13.6% useful info and so on
```

```python
pca.n_components_ # it gives how many columns you got
```

```python
Pxtr , Pxts , Pytr , Pyts = train_test_split( x_pca , y , test_size=0.2 , random_state=30 )
```

```python
len(Pxtr)
```

```python
lr2 = LogisticRegression( max_iter=10000 )
lr2.fit(Pxtr , Pytr )
```

```python
lr2.score( Pxts , Pyts )
```

```python
# let us consider n_components=2 , only 2 useful columns

pca = PCA( n_components=2 )
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
Pxtr , Pxts , Pytr , Pyts = train_test_split( x_pca , y , test_size=0.2 , random_state=30 )

lr2 = LogisticRegression( max_iter=10000 )
lr2.fit(Pxtr , Pytr )

lr2.score( Pxts , Pyts )
```
