```python
# handle missing data                                     Tut-5 : codeBasics

# 1. fillna
# 2. dropna
# 3. interpolate
```

```python
df1 = pd.read_csv("weather_data_2.csv")
df1
```

```python
type(df1.day[0])
```

```python
df1 = pd.read_csv("weather_data_2.csv", parse_dates=["day"])    #changes day type from 'str' to 'date'
```

```python
type(df1.day[0])
```

```python
df1.set_index('day',inplace=True )  # inplace=True : modifies the orginal dataset
                                    # this function sets day as indexing
```

```python
df1
```

```python
#fillna :

new_df1 = df1.fillna(0)        #it replace ,NaN values with 0
new_df1
```

```python
new_df1 = df1.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})

new_df1
```

```python
new_df1 = df1.fillna(method = "ffill")      # if NaN, carry forward prv day value
new_df1                                     # ffill - forward fill
                                            # bfill - backward fill
```

```python
new_df1 = df1.fillna(method = "ffill" , axis="columns")  #copies values horizontally
new_df1
```

```python
#interpolate :

new_df1 = df1.interpolate()
new_df1
```

```python
#dropna :

new_df = df1.dropna()
new_df
```

```python
new_df= df1.dropna(how="all")       # a row with all NaN values is completely removed
new_df
```

```python
new_df= df1.dropna(thresh=1)       #if there is atleast 1 valid value, don't drop it
new_df
```

```python
new_df= df1.dropna(thresh=2)
new_df
```
